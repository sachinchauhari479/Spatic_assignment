```python
import pandas as pd
from haversine import haversine
from difflib import SequenceMatcher

# Load the dataset into a pandas dataframe
df = pd.read_csv('/content/drive/MyDrive/assinment/assignment_data.csv')
```

```python
# Define a function to calculate string similarity
def is_similar(str1, str2):
    return SequenceMatcher(None, str1, str2).ratio() > 0.8 and abs(len(str1) - len(str2)) <= 5
```

```python
# Create a list to store the matches
matches = []

# Iterate over each entry in the dataframe
for i, row1 in df.iterrows():
    # Find all entries that are within 200 meters of this entry
    nearby_entries = df[df.apply(lambda row2: haversine((row1['latitude'], row1['longitude']), (row2['latitude'], row2['longitude'])) < 0.2, axis=1)]
    
    # Check if the names of the nearby entries are similar
    is_similar_entry = False
    for j, row2 in nearby_entries.iterrows():
        if i != j and is_similar(row1['name'], row2['name']):
            # The names are similar and the entries are within 200 meters of each other, so mark the entry as similar
            is_similar_entry = True
            break
    
    # Add the information about this entry and whether it is similar to the matches list
    matches.append({'name': row1['name'], 'latitude': row1['latitude'], 'longitude': row1['longitude'], 'is_similar': int(is_similar_entry)})


```

```python
# Convert the matches list to a pandas dataframe
matches_df = pd.DataFrame(matches)
```
